```python
import os
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '2'

print(tf.test.gpu_device_name())

import tensorflow as tf
physical_devices = tf.config.list_physical_devices('GPU')
tf.config.experimental.set_memory_growth(physical_devices[0], True)
```

```python
#initialization of Tensors
x = tf.constant(4)
print(x)

print ("\n")

x = tf.constant(4.0)
print(x)

print ("\n")


x = tf.constant(4, shape = (1,1))
print(x)

print ("\n")

x = tf.constant(4, shape = (1,1), dtype = tf.float32)
print(x)

print ("\n")

# Create an array - a 2-dim tensor
x = tf.constant([[1,2,3] , [4,5,6]])
print(x)

print ("\n")

x = tf.ones((3,3))
print(x)

print ("\n")

x = tf.zeros((2, 3))
print(x)

print ("\n")

x = tf.eye(3)   # identity matrix
print(x)

print ("\n")

x = tf.random.normal((3,3), mean = 0, stddev = 1)
print(x)

print ("\n")

x = tf.random.uniform((1, 3), minval = 0, maxval = 1)
print(x)

print ("\n")

x = tf.range(9)
print(x)

print ("\n")

x = tf.range(start = 1, limit = 10, delta = 2) # delta = step
print(x)

print ("\n")

x = tf.cast(x, dtype = tf.float64)
print(x)
```

```python
#Mathematical Operations
x = tf.constant([1, 2, 3])
y = tf.constant([9, 8, 7])

z = tf.add(x, y) # element wise adding
print(z)

print("\n")

z = x + y  # element wise adding
print(z)

print("\n")

z = tf.subtract(x, y)   # element wise subtraction
print(z)

print("\n")

z = x - y    # element wise subtraction
print(z)

print("\n")

z = tf.divide(x, y) # element wise division
print(z)

print("\n")

z = x / y # element wise division
print(z)

print("\n")

z = tf.multiply(x, y) # element wise multiplication
print(z)

print("\n")

z = x * y  # element wise multiplication
print(z)

print("\n")

z = tf.tensordot(x, y, axes = 1)  # dot product
print(z)

print("\n")

z = tf.reduce_sum(x * y, axis = 0) # dot product
print(z)

print("\n")

z = x ** 5  # element wise exponentiation
print(z)

print("\n")

x = tf.random.normal((2, 3))
y = tf.random.normal((3, 4))

z = tf.matmul(x, y) # matrix multiplication

print(z)

print("\n")

z = x @ y   # matrix multiplication

print(z)
```

```python
#Indexing
x = tf.constant([0, 1, 1, 2, 3, 1, 2, 3])

print('print elements of the tensor:')
print('-----------------------------\n')
print(x[:])
print(x[1:])
print(x[1:3])
print(x[::2])
print(x[::-1])

print("\n")

print('gather some specific values in the tensor to create a new tensor:')
print('-----------------------------------------------------------------\n')
indices = tf.constant([0, 3])
x_ind = tf.gather(x, indices)
print(x_ind)

print("\n")

print("Indexing a matrix:")
print("------------------\n")
x = tf.constant([[1, 2], [3, 4], [5, 6]])

print("print out the elements of the 3 x 2 matrix")
print(x)

print("\n")

print("get the first row with all of the elements:")
print(x[0])

print("\n")

print("get the first row with all of the elements:")
print(x[0,:])

print("\n")

print("print out the first two rows:")
print(x[1:])

```

```python
#Reshaping
x = tf.range(9)
print(x)

print("\n")

print("reshape x to a 3x3 matrix:")
x = tf.reshape(x, (3 ,3))
print(x)

print("\n")

print("transpose x:")
x = tf.transpose(x)
print(x)

print("\n")

print("Another way to transpose x:")
x = tf.range(9)
x = tf.reshape(x, (3 ,3))
x = tf.transpose(x, perm = [1, 0])
print(x)
```
